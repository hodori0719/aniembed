```python
import requests
from requests.structures import CaseInsensitiveDict

import pandas as pd

import json
import time
```

```python
# API auth info. Note the client-id is private, and the code will not work without your own
headers = CaseInsensitiveDict()
headers["X-MAL-CLIENT-ID"] = # Your client ID here

# API call settings
URL_USER_INFO = "https://api.myanimelist.net/v2/users/"
# the MAL API limits the number of anime calls at once to 1000
LIMIT = "&limit=1000"
OFFSET = "&offset="
# The id, title, and main_picture fields are included in calls by default.
FIELDS = "/animelist?sort=list_score&fields=list_status"
```

```python
anime_dict = {}
index = 0

with open("animelist.txt", 'r') as infile:
    for line in infile:
        anime_id = int(line.strip())
        anime_dict[anime_id] = index
        index += 1
    
usernames = open("users.txt", 'r')
```

```python
dataframe = pd.DataFrame(columns=anime_dict.keys())
```

```python
with open("users.txt", 'r') as infile:
    for line in infile:
        username = line.strip()
        response = requests.get(URL_USER_INFO + username + FIELDS + LIMIT, headers = headers)
        if response.status_code == 200:
            responsedict = json.loads(response.text)['data']
            for entry in responsedict:
                if int(entry['node']['id']) in anime_dict:
                    dataframe.at[username, int(entry['node']['id'])] = entry['list_status']['score']
        time.sleep(0.6)
```
